<h2><font color="orange">1) Retrieve non-hit songs from LastFM</font></h2>

```python
#Set the parameters to LastFM API calls. Method and Limit are empty so we can adjust them to specific calls\
#like get_top_tracks or track.getInfo
results_df = pd.DataFrame({
    "name": [],
    "duration": [],
    "mbid": [],
    "url": [],
    "streamable": [],
    "artist": [],
    "image": [],
    "@attr": []
})

#API call for first 1000 songs using tag.getTopTracks
for i in range(3, 13):
   
    page = i
    
    params={
    "api_key": api_key,
    "format": "json",
    "method":"tag.gettoptracks",
    "limit": 1000,
    "page": page
}
    base_url= "http://ws.audioscrobbler.com/2.0/?"
    example_url= "/2.0/?method=tag.gettoptracks&tag=disco&api_key=YOUR_API_KEY&format=json"    #taken from website
    
    print(f'Retrieving info for page:{i}')
    response=requests.get(base_url, params=params).json()

    temp_df=pd.DataFrame(response["tracks"]["track"])
    results_df=results_df.append(temp_df, ignore_index=True, sort=False)
    
results_df
```

```python
#drop duplicates
results_df.drop_duplicates(subset='name', keep='first', inplace=True)

results_df.head()
```

<h2><font color="orange">2) Retrieving Spotify hits and song features</font></h2>

```python
#Spotipy
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from config import spotify_client,spotify_secret
```

```python
#Spotify client authorization flow
client_credentials_manager = SpotifyClientCredentials(client_id=spotify_client, client_secret=spotify_secret) 
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
#Read in LastFM csv file
csv = "clean_last_fm_df.csv"
unnamed_data = pd.read_csv(csv)
unnamed_data.head()
```

```python
#Drop unnamed column
song_df = unnamed_data.drop(["Unnamed: 0"],axis=1)
song_df.head()
```

```python
#Track name array
tracks = song_df["name"]

# Empty lists to hold API result data
track_name = []
track_id = []
explicit = []
popularity = []

#Find Spotify id for each song in the LastFM dataset
for track in range(len(tracks)):
    try:
        #Create a print log for each country
        print(f"Retrieving information {track} | {tracks[track]}")
        
        #Get spotify id by searching track names
        track_results = sp.search(q=tracks[track], type="track",limit=1)
        
        #Append spotify id and some audio features to empty lists
        for i, t in enumerate(track_results["tracks"]["items"]):
            track_name.append(t["name"]) 
            track_id.append(t["id"])
            explicit.append(t["explicit"])
            popularity.append(t["popularity"])
        
    except ValueError as e:
        print(f"Missing {track} | {tracks[track]}")
```

```python
# Empty audio features list
features = []

#Pull audio features using Spotify ids
for t in range(len(track_id)):
    try:
        #Create a print log for each country
        print(f"Retrieving information {t} | {track_id[t]}")
        
        #Get track audio features and append to empty lists
        features.append(sp.audio_features(track_id[t]))
        
    except ValueError as e:
        print(f"Missing {track} | {tracks[track]}")
```

```python
#Create a df with all retrieved features
features_df=pd.DataFrame(features, columns={"Features"})
# features_df.to_csv("features_df.csv")
features_df.head()
```

```python
#Create a df with all ids
id_df = pd.DataFrame({"Name":track_name,"Spotify ID":track_id,"Explicit":explicit,"Spotify Popularity":popularity})
# id_df.to_csv("track_id.csv")
id_df.head()
```

```python
#Add separate columns for each feature to the id_df 

features_df1=id_df    # <- this is just to create a copy of the id_df so that we don't mess up with the retrieved ids
feature_columns=[
    "danceability",
    "energy",
    "key",
    "loudness",
    "mode",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence",
    "tempo",
    "type",
    "id",
    "uri",
    "track_href",
    "analysis_url",
    "duration_ms",
    "time_signature"
]

#add an empty column for each feature
for column in feature_columns:
    features_df1[column]=""
features_df1.head()
```

```python
#Populate the new columns of the df with retrieved features
for index, row in features_df1.iterrows():
    print(f'Adding features for index:{index}')
    features_df1["danceability"][index]=features_df["Features"][index]["danceability"]
    features_df1["energy"][index]=features_df["Features"][index]["energy"]
    features_df1["key"][index]=features_df["Features"][index]["key"]
    features_df1["loudness"][index]=features_df["Features"][index]["loudness"]
    features_df1["mode"][index]=features_df["Features"][index]["mode"]
    features_df1["speechiness"][index]=features_df["Features"][index]["speechiness"]
    features_df1["acousticness"][index]=features_df["Features"][index]["acousticness"]
    features_df1["instrumentalness"][index]=features_df["Features"][index]["instrumentalness"]
    features_df1["liveness"][index]=features_df["Features"][index]["liveness"]
    features_df1["valence"][index]=features_df["Features"][index]["valence"]
    features_df1["tempo"][index]=features_df["Features"][index]["tempo"]
    features_df1["type"][index]=features_df["Features"][index]["type"]
    features_df1["id"][index]=features_df["Features"][index]["id"]
    features_df1["uri"][index]=features_df["Features"][index]["uri"]
    features_df1["track_href"][index]=features_df["Features"][index]["track_href"]
    features_df1["analysis_url"][index]=features_df["Features"][index]["analysis_url"]
    features_df1["duration_ms"][index]=features_df["Features"][index]["duration_ms"]
    features_df1["time_signature"][index]=features_df["Features"][index]["time_signature"]
```

```python
#Add a column "hit" with zeros as label for "no hit" (hit=1 vs no hit=0). Save df
features_df1["hit"]=0
# features_df1.to_csv("tracks+id+features.csv")
features_df1.head()
```
